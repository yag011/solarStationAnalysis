```python
import pandas as df
import numpy as np
from datetime import date
import matplotlib.pyplot as plt
pd = df.read_csv('north-transfer-station-solar-panels.csv', index_col=0, parse_dates=True)
pd.dropna(axis = 0, how ='any', inplace=True)
Q1 = pd.quantile(0.25)
Q3 = pd.quantile(0.75)
IQR = Q3 - Q1
pd = pd[~((pd < (Q1 - 1.5 * IQR)) |(pd > (Q3 + 1.5 * IQR))).any(axis=1)]
pd.plot(title = 'Inverters 1-4 over Time',legend = False, sharex='col', sharey='row', subplots=True)
```

The line plots below are designed to provide a graphical presentation of the similarities among the four inverters and seasonal variability.  

```python
spr1=[]
sum1=[]
fall1=[]
win1=[]
spr2=[]
sum2=[]
fall2=[]
win2=[]
spr3=[]
sum3=[]
fall3=[]
win3=[]
spr4=[]
sum4=[]
fall4=[]
win4=[]
for i in newdf.itertuples():
    word=i.Date
    word=word.split('-')
    if word[1]=='Mar' or word[1]=='Apr' or word[1]=='May':
        spr1.append(i.Inverter1)
        spr2.append(i.Inverter2)
        spr3.append(i.Inverter3)
        spr4.append(i.Inverter4)
    if word[1]=='Jun' or word[1]=='Jul' or word[1]=='Aug':
        sum1.append(i.Inverter1)
        sum2.append(i.Inverter2)
        sum3.append(i.Inverter3)
        sum4.append(i.Inverter4)
    if word[1]=='Sep' or word[1]=='Oct' or word[1]=='Nov':
        fall1.append(i.Inverter1)
        fall2.append(i.Inverter2)
        fall3.append(i.Inverter1)
        fall4.append(i.Inverter1)
    if word[1]=='Dec' or word[1]=='Jan' or word[1]=='Feb':
        win1.append(i.Inverter1)
        win2.append(i.Inverter2)
        win3.append(i.Inverter3)
        win4.append(i.Inverter4)
```

grouping the data into seasons 

```python
import scipy.stats as stats
f_value, p_value = stats.f_oneway(spr1,spr2,spr3,spr4)
print(f_value, "   ", p_value)
f_value1, p_value1 = stats.f_oneway(sum1,sum2,sum3,sum4)
print(f_value1, "   ", p_value1)
f_value2, p_value2 = stats.f_oneway(fall1,fall2,fall3,fall4)
print(f_value2, "   ", p_value2)
f_value3, p_value3 = stats.f_oneway(win1,win2,win3,win4)
print(f_value3, "   ", p_value3)
```

Four one-way anova tests  has been performed on each season for all four inverters.
